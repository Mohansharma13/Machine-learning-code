```python
from sklearn.datasets._samples_generator import make_circles
```

```python
x,y=make_circles(200,noise=0.1,factor=0.1)
```

```python
x.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[:,0],x[:,-1],c=y)
```

```python
from sklearn.svm import SVC
```

```python
svcclasifier=SVC(kernel="linear")
svcclasifier.fit(x,y)
```

```python
svcclasifier.score(x,y)
```

```python
def plot_decision_boundary(clf, X, y):
    import numpy as np
    # Create a mesh to plot the decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))

    # Predict the function value for the whole grid
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the contours
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)

    # Plot the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap='bwr')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('SVM Decision Boundary')

    # Highlight the support vectors
    support_vectors = clf.support_vectors_
    plt.scatter(support_vectors[:, 0], support_vectors[:, 1], facecolors='none', edgecolors='k', s=100, linewidths=1.5, label='Support Vectors')
    plt.legend()
    plt.show()
```

```python
plot_decision_boundary(svcclasifier,x,y)
```

```python
svcclasifier.score(x,y) # poor desion boundary
```

# using kernal trick

```python
def plot_3d_plot(X, y):
    import numpy as np
    r = np.exp(-(X ** 2).sum(1))
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:, 0], X[:, 1], r, c=y, s=100, cmap='bwr')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('y')
    return ax
```

```python
plot_3d_plot(x,y)
```

```python
# as we can see kernal classify the data
```
