```python
dict = df2.set_index('id')['name'].to_dict()
```

```python
import ast
def id_to_name(lst):
    my_list = ast.literal_eval(lst)
    new_list=[]
    for i in my_list:
        if i in dict:
            new_list.append(dict[i])
    return new_list
```

```python
df1['genre_ids']=df1['genre_ids'].apply(id_to_name)
```
