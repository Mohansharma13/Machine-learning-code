```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([1.,2.])
y_tarin=np.array([300.,500.])
```

```python
#function to graduent desent
def compute_gradient(x,y,w,b):
    m=x.shape[0]
    dj_dw=0
    dj_db=0
    for i in range(m):
        f_wi=w*x[i]+b
        dj_dw=(f_wi-y[i])*x[i]
        dj_db=(f_wi-y[i])
        dj_db+=dj_db
        dj_dw+=dj_dw
        
    dj_dw=dj_dw/m
    dj_db=dj_db/m
    return dj_db,dj_dw
```

```python
def gradient_decent(x,y,w_in,b_in,alpha,num_iter):
    b=b_in
    w=w_in
    for i in range(num_iter):
        dj_db , dj_dw=compute_gradient(x,y,w,b)
        w=w-alpha*dj_dw
        b=b-alpha*dj_db
    return w,b
```

```python
w_init=0
b_init=0
it=10000
tmpaplha=0.001
w_final,b_final=gradient_decent(x_train , y_tarin, w_init, b_init, tmpaplha, it)
```

```python
plt.scatter(x_train,y_tarin)
y_hat=w_final*x_train+b_final

plt.plot(x_train,y_hat)
```
